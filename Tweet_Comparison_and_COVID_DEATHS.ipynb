# Barplot

```python
x_nj=nj_covid_data['date']
y_nj=nj_covid_data['death']
x1_fl=fl_covid_data['date']
y1_fl=fl_covid_data['death']

plt.subplot(2,1,1)
pop = plt.bar(x_nj, y_nj)
plt.ylabel('Death In NJ')
plt.xlabel("Date")
plt.xticks(rotation='vertical')
plt.subplot(2,1,2)

gdp =plt.bar(x1_fl,y1_fl)
plt.ylabel('Death In FL')

plt.xticks(rotation='vertical')

plt.show()
```

```python
rawtweets = pd.read_csv('COVID.csv', usecols=['Tweet Content', 'Tweet Location'])
rawtweets
```

```python
rawtweets2 = rawtweets.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
raw_FL = rawtweets2[rawtweets2['Tweet Location'].str.contains('FL',na=False)]
raw_FL
```

```python
raw_Florida = rawtweets2[rawtweets2['Tweet Location'].str.contains('Florida',na=False)]
raw_Florida
```

```python
raw_Florida = rawtweets2[rawtweets2['Tweet Location'].str.contains('Florida',na=False)]
raw_Florida.drop([899])
```

```python
raw_FLORIDA = rawtweets2[rawtweets2['Tweet Location'].str.contains('FLORIDA',na=False)]
raw_FLORIDA
```

```python
raw_NJ = rawtweets2[rawtweets2['Tweet Location'].str.contains('NJ',na=False)]
raw_NJ
```

```python
raw_New_Jersey = rawtweets2[rawtweets2['Tweet Location'].str.contains('New Jersey',na=False)]
raw_New_Jersey
```

**DATAFRAME FOR NJ**

```python
frames_2 = [raw_New_Jersey, raw_NJ]
raw_nj_updated = pd.concat(frames_2)
raw_nj_updated
```

**DATAFRAME FOR FLORIDA**

```python
frames = [raw_FLORIDA, raw_Florida]
raw_fl_updated = pd.concat(frames)
raw_fl_updated
```

# Word Clouds

# WORD CLOUD NJ

```python
df= raw_nj_updated.copy()
df= df.reset_index(drop= True)
df
```

```python
from wordcloud import WordCloud 
data_pos = df['Tweet Content']
wc = WordCloud(max_words = 1000 , width = 1600 , height = 800,
              collocations=False).generate(" ".join(df['Tweet Content']))
plt.figure(figsize = (20,20))
plt.imshow(wc)
```
